```python
#importing libraries
import numpy as np
import pandas as pd

pd.set_option('mode.chained_assignment', None) #to get rid of a warning
```

```python
#specifying columns to use
usecols1 = ['CustomerID', 'CustomerName', 'CustomerCategoryID', 'BuyingGroupID', 'PrimaryContactPersonID']
usecols2 = ['CustomerCategoryID', 'CustomerCategoryName']
usecols3 = ['PersonID', 'FullName']
usecols4 = ['CustomerID', 'TransactionDate', 'TransactionAmount', 'IsFinalized']
usecols5 = ['SupplierID', 'SupplierName', 'SupplierCategoryID', 'PrimaryContactPersonID']
usecols6 = ['SupplierCategoryID', 'SupplierCategoryName']
usecols7 = ['SupplierID', 'PurchaseOrderID', 'TransactionAmount', 'IsFinalized']
usecols8 = ['Admin2', 'Province_State', 'Country_Region', 'Lat', 'Long_', 'Combined_Key', 'Date', 'Confirmed', 'Deaths']
usecols9 = ['Date', 'Temperature (° F) - Avg', 'Dew Point (° F) - Avg', 'Humidity (%) - Avg', 'Wind Speed (mph) - Avg', 
           'Pressure (Hg) - Avg', 'Precipitation (in) Total']
    
#load data
cs = pd.read_csv("CUSTOMERS.csv", usecols=usecols1)
csc = pd.read_csv("CUSTOMER_CATEGORY.csv", usecols=usecols2)
ps = pd.read_csv("PERSONS.csv", usecols=usecols3)
tr = pd.read_csv("TRANSACTIONS.csv", usecols=usecols4)
sp = pd.read_csv("SUPPLIERS.csv", usecols=usecols5)
spc = pd.read_csv("SUPPLIER_CATEGORY.csv", usecols=usecols6)
spt = pd.read_csv("SUPPLIER_TRANSACTION.csv", usecols=usecols7)
cov = pd.read_csv("Usa_county_covid 19.csv", usecols=usecols8)
wt = pd.read_csv("Weather.csv", encoding='latin1', usecols=usecols9)  #I specified the encoding because "utf-8" could not decode its special characters
```

```python
#I will rename some columns to clean up the data a little bit
cs.rename(columns={"PrimaryContactPersonID":"PersonID"}, inplace=True)
csc.rename(columns={"CustomerCategoryName":"Goods_Category"}, inplace=True)
ps.rename(columns={"FullName":"Agent_Name"}, inplace=True)
tr.rename(columns={"TransactionDate":"Date"}, inplace=True)
sp.rename(columns={"PrimaryContactPersonID":"PersonID"}, inplace=True)
spc.rename(columns={"SupplierCategoryName":"Goods_Category"}, inplace=True)
```

### I want to merge all the sales tables together, to make it easier for analysis

```python
sales1=cs.merge(csc,on='CustomerCategoryID', how="left") #I merged on the left because I would want to retain all the data in the first dataframe and also use the information on the second table
sales2=ps.merge(sales1,on='PersonID', how="left")
sales=tr.merge(sales2,on='CustomerID', how="left")
sales
```

### I also want to do the same for the suppliers tables

```python
suppliers1=sp.merge(spc,on='SupplierCategoryID', how="left") #I merged on the left because I would want to retain all the data in the first dataframe and also use the information on the second table
suppliers=spt.merge(suppliers1,on='SupplierID', how="left")
suppliers
```

## Data Inspection and Wrangling

### Sales

```python
#check missing values
sales.isnull().sum()
```

#### We have 50530 missing values in BuyingGroupID, but from the excel file that explained the variables, the missing values is indicating that the customer(s) is an individual and not in a group. The analysis we are going to do will be based on the individual customers, so we have to remove the data of the transactions done in group(s).
#### Also, IsFinalized '0' values has to be dropped too because it indicates the transactions that have not been finalized.

```python
sales['BuyingGroupID'].unique()
```

```python
sales.replace({'BuyingGroupID': {np.NaN: 10}}, inplace = True) #I replaced it with 10 (you can use any number aside 1 or 2) to change it to notnull value
sales.replace({'BuyingGroupID': {1.0: np.NaN}}, inplace = True) #I changed it to null value so that I can easily drop it
sales.replace({'BuyingGroupID': {2.0: np.NaN}}, inplace = True)
sales.replace({'IsFinalized': {0: np.NaN}}, inplace = True)
sales
```

```python
sales.isnull().sum()
```

#### Now, we can go ahead and drop all the rows of the missing values

```python
sales = sales.dropna() 
sales
```

```python
#checking data types
sales.dtypes
```

```python
#They are okay except 'Date', let's change it to date type
sales['Date'] = pd.to_datetime(sales['Date'])
sales.dtypes
```

```python
#let us split the 'TransactionDate' to have month, quarter and year column s.loc[[1, 2]]
sales['Year']= sales['Date'].dt.year
sales['Quarter']= sales['Date'].dt.quarter
sales['Month']= sales['Date'].dt.month
sales
```

```python
#Now we can export the sales data to be explored in a seperate notebook
sales.to_csv('Sales.csv',index=False)
```
