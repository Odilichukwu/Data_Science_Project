```python
# Setup

# common:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load data
sales = pd.read_csv("Sales.csv") 
supplier = pd.read_csv("DSuppliers.csv") 
wt = pd.read_csv("DWeather.csv")
```

## Sales and Suppliers

```python
#to visualize the total amount paid to the suppliers
spn=supplier.groupby('SupplierName').TransactionAmount.sum().reset_index()
plt.figure(figsize=(13, 4))
ax = sns.barplot(x='SupplierName', y="TransactionAmount", data=spn)  
```

```python
#to visualize the total amount paid for the supply of the goods category
catg=supplier.groupby('Goods_Category').TransactionAmount.sum().reset_index()
plt.figure(figsize=(10, 4))
ax = sns.barplot(x='Goods_Category', y="TransactionAmount", data=catg)    
```

```python
#to visualize the total amount sold by goods category
cat=sales.groupby('Goods_Category').TransactionAmount.sum().reset_index()
plt.figure(figsize=(10, 4))
ax = sns.barplot(x='Goods_Category', y="TransactionAmount", data=cat)    
```

From the above two category charts, the novelty shop sells the most but is also the least purchase, while corporate which probably belongs to clothing supplies, sells the least. 
The company need to evaluate thier purchase options so as not to tie down their capital purchasing the least sold items.

```python
df=supplier.groupby(['SupplierName', 'Goods_Category']).TransactionAmount.sum().sort_values(ascending=False).reset_index()
total = df["TransactionAmount"].sum()
df["Supplier transaction in %"] = round(df["TransactionAmount"] / total * 100, 4)
df
```

```python
df.columns=['SupplierName', 'Goods_Category',"sum", "Supplier transaction in %"]
plt.figure(figsize=(13, 6))
ax = sns.barplot(x='SupplierName', y="sum", hue='Goods_Category', data=df)
    
```

## Sales and Weather

```python
#to create a new dataframe that has only the month and temperature columns
wth = pd.DataFrame(columns=['Month', 'Temp'], data=wt[["Month","Temperature (Â° F) - Avg"]].values)
wtt = wth.groupby('Month').Temp.mean().reset_index()
wtt
```

```python
#to regroup the thw dataframe by goods category
gift=sales[sales.Goods_Category=='Gift Store'].reset_index(drop=True)
novelty=sales[sales.Goods_Category=='Novelty Shop'].reset_index(drop=True)
superm=sales[sales.Goods_Category=='Supermarket'].reset_index(drop=True)
comp=sales[sales.Goods_Category=='Computer Store'].reset_index(drop=True)
corp=sales[sales.Goods_Category=='Corporate'].reset_index(drop=True)
```

```python
#to obtain total average of goods sold in a month by their categories
gift=gift.groupby('Month').TransactionAmount.mean().reset_index()
novelty=novelty.groupby('Month').TransactionAmount.mean().reset_index()
superm=superm.groupby('Month').TransactionAmount.mean().reset_index()
comp=comp.groupby('Month').TransactionAmount.mean().reset_index()
corp=corp.groupby('Month').TransactionAmount.mean().reset_index()

gift.columns=['Month',"sum"] 
novelty.columns=['Month',"sum"] 
superm.columns=['Month',"sum"] 
comp.columns=['Month',"sum"] 
corp.columns=['Month',"sum"] 

#to scale down the data to enable visualizing with the temperature values
gift['sum'] = gift['sum'] / 10000
novelty['sum'] = novelty['sum'] / 10000
superm['sum'] = superm['sum'] / 10000
comp['sum'] = comp['sum'] / 10000
corp['sum'] = corp['sum'] / 10000
```

```python
#to visualize the trend of amount of goods by category sold and the temperature
#Gift store
vizg=pd.merge(wtt,gift,on="Month",how="inner")
vizgg = vizg.set_index('Month')
vizgg.plot(figsize = (13, 5))
```

```python
#Novelty shop
vizn=pd.merge(wtt,novelty,on="Month",how="inner")
viznn = vizn.set_index('Month')
viznn.plot(figsize = (13, 5))
```

```python
#supermarket
vizs=pd.merge(wtt,superm,on="Month",how="inner")
vizss = vizs.set_index('Month')
vizss.plot(figsize = (13, 5))
```

```python
#computer store
vizco=pd.merge(wtt,comp,on="Month",how="inner")
vizcoo = vizco.set_index('Month')
vizcoo.plot(figsize = (13, 5))
```

```python
#corporate
vizcp=pd.merge(wtt,corp,on="Month",how="inner")
vizcpp = vizcp.set_index('Month')
vizcpp.plot(figsize = (13, 5))
```
